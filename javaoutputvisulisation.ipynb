```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from specified file paths
velocities = pd.read_csv(r"C:\Users\shreevats\OneDrive\Desktop\Project\velocities.csv", header=None, names=["Velocity"])
momenta = pd.read_csv(r"C:\Users\shreevats\OneDrive\Desktop\Project\momenta.csv", header=None, names=["Momentum"])
kinetic_energies = pd.read_csv(r"C:\Users\shreevats\OneDrive\Desktop\Project\kinetic_energies.csv", header=None, names=["Kinetic Energy"])

# Function to plot histograms
def plot_histogram(data, column_name, title, xlabel):
    plt.figure(figsize=(8, 6))
    plt.hist(data[column_name], bins=30, edgecolor='k', alpha=0.7)
    plt.title(f"{title} (Histogram)")
    plt.xlabel(xlabel)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# Function to plot box plot
def plot_boxplot(data, column_name, title):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[column_name])
    plt.title(f"{title} (Box Plot)")
    plt.xlabel(column_name)
    plt.grid(True)
    plt.show()

# Function to plot scatter plot
def plot_scatter(data1, column1, data2, column2, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(data1[column1], data2[column2], alpha=0.7, edgecolor='k')
    plt.title(f"{title} (Scatter Plot)")
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.grid(True)
    plt.show()

# Function to plot KDE plot
def plot_kde(data, column_name, title):
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data[column_name], shade=True)
    plt.title(f"{title} (KDE Plot)")
    plt.xlabel(column_name)
    plt.grid(True)
    plt.show()

# Visualizations for Velocity
plot_histogram(velocities, "Velocity", "Velocity Magnitude Distribution", "Velocity")
plot_boxplot(velocities, "Velocity", "Velocity Magnitude Distribution")
plot_kde(velocities, "Velocity", "Velocity Magnitude Distribution")

# Visualizations for Momentum
plot_histogram(momenta, "Momentum", "Momentum Distribution", "Momentum")
plot_boxplot(momenta, "Momentum", "Momentum Distribution")
plot_kde(momenta, "Momentum", "Momentum Distribution")

# Visualizations for Kinetic Energy
plot_histogram(kinetic_energies, "Kinetic Energy", "Kinetic Energy Distribution", "Kinetic Energy")
plot_boxplot(kinetic_energies, "Kinetic Energy", "Kinetic Energy Distribution")
plot_kde(kinetic_energies, "Kinetic Energy", "Kinetic Energy Distribution")

# Scatter plot for comparison between Momentum and Kinetic Energy
plot_scatter(momenta, "Momentum", kinetic_energies, "Kinetic Energy", "Momentum vs Kinetic Energy")
```
